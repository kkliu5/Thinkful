```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('Downloads/fivethirtyeight-comic-characters-dataset/marvel-wikia-data.csv')

df['male_count'] =  np.where(df['SEX']== 'Male Characters',1,0)
df['female_count'] =  np.where(df['SEX']== 'Female Characters',1,0)

plt.scatter(
    x=df['Year'],
    y=df['APPEARANCES'],
    color='green',
    marker='x',
    s=20
)
plt.ylabel('Appearances')
plt.xlabel('Year Appeared')
plt.title('Year Introduced Compared Against Appearance Frequency')


plt.tight_layout()
plt.show()
```

This graph shows frequently appeared super heros throughout the years. You can see super heros created from 1960-1970 were very popular since they appeared frequently.

```python
var = df.groupby('SEX').APPEARANCES.sum() 

plt.subplot(1, 2, 1)
var.plot(kind='bar') 

plt.ylabel('Appearances')
plt.title('Appearances by Sex')

plt.subplot(1, 2, 2)
df['SEX'].value_counts().plot(kind='bar')

plt.ylabel('Frequency')
plt.xlabel('Sex')
plt.title('Total Number of Heros by Sex')

plt.tight_layout()
plt.show()
```

These two graphs shows that Marvel comics are dominated by male superheros. In the first chart you see the appearances of each superhero split by sex, which shows that male superheros appear more often than female superheros by double. The second graph shows how many superheros are in the Marvel universe split by sex, which is signifacntly dominated by male heros by almost quadruple the amount. 

```python
plt.hist(
    df['Year'],
    color='red'
)

plt.ylabel('Frequency')
plt.xlabel('Year')
plt.title('Total Number of Created Heros by Year')

plt.tight_layout()
plt.show()
```

This graph shows that new superheros were created at a rapid quantity from the years 1990-2010 compared to the earlier years from 1940-1960.
